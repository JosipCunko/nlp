```python
import nltk
from nltk.tokenize import word_tokenize as wordTokenize
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bbc = pd.read_csv("bbc_news.csv")
```

```python
bbc.head()
```

```python
bbc.info()
```

```python
titles = pd.DataFrame(bbc["title"])
titles.head(6)
```

```python
titles["lowercase"] = titles["title"].str.lower()
```

```python
enStopwords = stopwords.words("english")
enStopwords.remove("did")
enStopwords.remove("not")
enStopwords.append("go")
```

```python
titles["noStopwords"] = titles["lowercase"].apply(lambda x: " ".join(word for word in x.split(" ") if word not in enStopwords))
```

```python
titles["noPunct"] = titles["noStopwords"].apply(lambda x: re.sub(r"[^\w\s]", "", x))
```

```python
titles["tokenized"] = titles["noPunct"].apply(lambda x: wordTokenize(x))
titles["rawTokens"] = titles.apply(lambda x: wordTokenize(x["title"]), axis =1)
```

```python
lemmatizer = WordNetLemmatizer()
titles["lemmatized"] = titles["tokenized"].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
titles.head()
```

```python
rawTokens = sum(titles["rawTokens"], [])
tokens = sum(titles["lemmatized"], [])
```

```python
nlp = spacy.load("en_core_web_sm")
spacyDoc = nlp(" ".join(rawTokens))
#spacyDoc
```

```python
posDf = pd.DataFrame(columns=["tokens", "pos_tag"])
```

```python
#highly inefficient
for token in spacyDoc:
    posDf = pd.concat([posDf,
                      pd.DataFrame.from_records([{"token": token.text, "pos_tag": token.pos_}])], ignore_index = True)
```

```python
posDfCounts = posDf.groupby(["token", "pos_tag"]).size().reset_index(name="counts").sort_values(by="counts", ascending=False)
```

```python
nouns = posDfCounts[posDfCounts.pos_tag == "NOUN"][0:10]
verbs = posDfCounts[posDfCounts.pos_tag == "VERB"][0:10]
verbs
```

```python
nerDf = pd.DataFrame(columns=["tokens", "ner_tag"])

for token in spacyDoc.ents:
    if pd.isna(token.label_) is False: #valid label?
        nerDf = pd.concat([nerDf, pd.DataFrame.from_records(
            [{"token": token.text, "ner_tag": token.label_}])], ignore_index = True)
```
