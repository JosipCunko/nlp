```python
import nltk #Natural Language Toolkit
nltk.download("stopwords") 
from nltk.corpus import stopwords
import re #regex

nltk.download("punkt_tab")
from nltk.tokenize import word_tokenize as wordTokenize, sent_tokenize as sentTokenize #tokenization

from nltk.stem import PorterStemmer #stemming
ps = PorterStemmer()

nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer #lemmatization
lemmatizer = WordNetLemmatizer()

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
data.info()
```

```python
data.head()
```

```python
data["lowercase"] = data["Review"].str.lower()
```

```python
enStopwords = stopwords.words("english")
enStopwords.remove("not")
data["noStopwords"] = data["lowercase"].apply(lambda review: " ".join([word for word in review.split(" ") if word not in enStopwords])) #execute callback for each element inside a specified column
```

```python
data["noPunctuation"] = data.apply(lambda review: re.sub(r"[^\w\s]", "" , review["noStopwords"]), axis = 1) #go row by row
data["noPunctuation"] = data.apply(lambda review: re.sub(r"[*]", "star" , review["noPunctuation"]), axis = 1)
```

```python
data["tokenized"] = data.apply(lambda review: wordTokenize(review["noPunctuation"]), axis =1)
```

```python
data["stemmed"] = data["tokenized"].apply(lambda tokens:[ps.stem(token) for token in tokens])
```

```python
data["lemmatized"] = data["tokenized"].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
cleanTokens = sum(data["lemmatized"], []) # .concat()
```

```python
unigrams = (pd.Series(nltk.ngrams(cleanTokens, 1))).value_counts()
biograms = (pd.Series(nltk.ngrams(cleanTokens, 2))).value_counts()
```
