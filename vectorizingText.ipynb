# bag of words model

```python
import pandas as pd
# transforms a collection of text documents into a matrix of token counts
# - it breaks the text into words(tokens) and counts how often each word appears
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
data = [' Most shark attacks occur about 10 feet from the beach since that is where the people are',
        'the efficiency with which he paired the socks in the drawer was quite admirable',
        'carol drank the blood as if she were a vampire',
        'giving directions that the mountains are to the west only works when you can see them',
        'the sign said there was road work ahead so he decided to speed up',
        'the gruff old man sat in the back of the bait shop grumbling to himself as he scooped out a handful of worms']
```

```python
countVec = CountVectorizer(binary=True) #words exists or not, only counts words
```

```python
countVecFit = countVec.fit_transform(data) # fit=> which unique words appear, transform => converts text into numbers and counts how often each word occurs
```

```python
countVecFit
```

```python
#row = doc, column = word, entry = word count
bagofwords = pd.DataFrame(countVecFit.toarray(), columns = countVec.get_feature_names_out())
bagofwords
```

```python
# term frequency-inverse document frequency
from sklearn.feature_extraction.text import TfidfVectorizer
# how important a word is, context, how often a word appears in a single doc(TF), how common it is across all docs(IDF)
# common words: lower score - dont destinguish on doc from another, higher score = more specific to a doc + importance
```

```python
tfidfVec = TfidfVectorizer()
```

```python
tfidfVecFit = tfidfVec.fit_transform(data)
```

```python
tfidfBag = pd.DataFrame(tfidfVecFit.toarray(), columns = tfidfVec.get_feature_names_out())
tfidfBag
```
