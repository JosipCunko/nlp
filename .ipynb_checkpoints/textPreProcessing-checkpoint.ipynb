### LOWERCASE and STOPWORDS

```python
import nltk
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
enStopwords = stopwords.words("english")
enStopwords.remove("did")
enStopwords.remove("not")
enStopwords.append("go")
print(enStopwords, end=" ")
```

```python
sentence = "Today I studided electronics and came to an conclusion that I am an genius"
```

```python
def removeStopwords(sentence):
    newSentence = ""
    for word in sentence.split(" "):
        if word not in enStopwords:
            newSentence+=word + " "
        
    return newSentence.lower()
```

```python
sentenceNoStopwords = " ".join(word for word in sentence.split() if word not in enStopwords).lower()
```

### TOKENIZATION
- breaking text into smaller units (tokens)
- easier for analysis

```python
nltk.download("punkt_tab")
from nltk.tokenize import word_tokenize as wordTokenize, sent_tokenize as sentTokenize
```

```python
sentences = "Transformer can increase the voltage. Tesla coil outputs only AC current"
sentTokenize(sentences)
```

```python
sentence2 = "Tesla's inventions changed the world"
wordTokenize(sentence2)
```

### STEMMING
- words reduced to thier base form
- reduces complexity


```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
connectTokens = ["connecting", "connected", "connect", "connects", "angrily"]
for t in connectTokens:
    print(t,": ", ps.stem(t))
```

### Lemmatization
- reduces a word to a meaningful base form

```python
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for t in ["likes", "better", "worse", "eggs", "angrily"]:
    print(t, ": ", lemmatizer.lemmatize(t))
```

### N Gram
- analyze the relationship between neighbouring words
- a sequence of N tokens
- unigram (a single word token)
- n > 3 is called a n-gram

```python
import pandas as pd # structured data, spreadsheet => Series(1D) or DataFrame(2D) table
import matplotlib.pyplot as plt # visualization
```

```python
tokens = ['the', 'rise', 'of', 'artificial', 'intelligence', 'has', 'led', 'to', 'significant', 'advancements', 'in', 'natural', 'language', 'processing', 'computer', 'vision', 'and', 'other', 'fields', 'machine', 'learning', 'algorithms', 'are', 'becoming', 'more', 'sophisticated', 'enabling', 'computers', 'to', 'perform', 'complex', 'tasks', 'that', 'were', 'once', 'thought', 'to', 'be', 'the', 'exclusive', 'domain', 'of', 'humans', 'with', 'the', 'advent', 'of', 'deep', 'learning', 'neural', 'networks', 'have', 'become', 'even', 'more', 'powerful', 'capable', 'of', 'processing', 'vast', 'amounts', 'of', 'data', 'and', 'learning', 'from', 'it', 'in', 'ways', 'that', 'were', 'not', 'possible', 'before', 'as', 'a', 'result', 'ai', 'is', 'increasingly', 'being', 'used', 'in', 'a', 'wide', 'range', 'of', 'industries', 'from', 'healthcare', 'to', 'finance', 'to', 'transportation', 'and', 'its', 'impact', 'is', 'only', 'set', 'to', 'grow', 'in', 'the', 'years', 'to', 'come']
print(tokens, end=" ")
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens, 1))).value_counts() # n = 1
print(unigrams)
```

```python
unigrams[:10].sort_values().plot.barh(color="lightsalmon", width= .9, figsize=(12,8))
plt.title("10 Most frequenlty occuring unigrams")
```

```python
biograms = (pd.Series(nltk.ngrams(tokens, 2))).value_counts() 
print(biograms[:10])
```

```python
trigrams = (pd.Series(nltk.ngrams(tokens, 3))).value_counts() 
print(trigrams[:10])
```
