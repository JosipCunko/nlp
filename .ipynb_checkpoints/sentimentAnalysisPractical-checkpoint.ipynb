```python
import pandas as pd
import numpy as np
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import transformers
from transformers import pipeline
```

```python
data = pd.read_csv("book_reviews_sample.csv")
data.head()
```

```python
data["reviewTextClean"] = data["reviewText"].str.lower()
data["reviewTextClean"] = data.apply(lambda x: re.sub(r"([^\w\s])", "", x["reviewTextClean"]), axis = 1)
```

```python
data.head()
```

```python
vaderSentiment = SentimentIntensityAnalyzer()
```

```python
data["vaderScore"] = data["reviewTextClean"].apply(lambda x:vaderSentiment.polarity_scores(x)["compound"])
```

```python
data.head()
```

```python
bins = [-1, -0.1, 0.1, 1]
labels  = ["negative", "neutral", "positive"]
data["vaderLabel"] = pd.cut(data["vaderScore"], bins, labels = labels)
```

```python
data.head()
```

```python
data["vaderLabel"].value_counts().plot.bar()
```

```python
transformerPipeline = pipeline("sentiment-analysis")
```

```python
transformerLabels = []
for review in data["reviewTextClean"].values:
    sentList = transformerPipeline(review)
    sentLabel = [sent["label"] for sent in sentList]
    transformerLabels.append(sentLabel)

data["transformerSentLabel"] = transformerLabels
```

```python
data["transformerSentLabel"].value_counts().plot.bar()
```
